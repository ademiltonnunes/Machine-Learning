#1. Import libraries

```python
import numpy as np
import pandas as pd
```

#2. Reading data

```python
data = pd.read_csv("data.csv")
```

#3. Describing data

```python
data.head()
```

#4. Filling missing data with median

```python
data["Working Experience(years)"].fillna(data["Working Experience(years)"].median(),inplace=True)
```

```python
data.head()
```

```python
data["Working Experience(years)"]=data["Working Experience(years)"].astype(int)
```

```python
data.head()
```

#5 Correlation between Number of Kids - Working Experience/ Number of Kids - Ages

```python
ageCorrelation:float  = data["Number of Kids"].corr(data["Age"])
print(ageCorrelation)
```

```python
WorkingExperinceCorrelation:float  =data["Working Experience(years)"].corr(data["Number of Kids"])
print(WorkingExperinceCorrelation)
```

```python
result:str = ""
correlation :float = 0

if ageCorrelation> WorkingExperinceCorrelation:
  result = "Age"
  correlation = ageCorrelation
else:
  result = "Working Experience(years)"
  correlation = WorkingExperinceCorrelation

print(f"{result} is more related to Number of Kids - {correlation} correlation")
```

6. One Hot Vector

```python
blood_types_encoded,categories=data["Blood Types"].factorize()
blood_types_encoded
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse=False)
blood_type_cat_1hot = encoder.fit_transform(blood_types_encoded.reshape(-1,1))
blood_type_cat_1hot
```

```python
data.head()
```

```python
one_hot=pd.get_dummies(data,columns=["Blood Types"],drop_first=False,prefix='',prefix_sep='')
one_hot
```

```python
data=data.drop("Blood Types",axis=1)
data
```

```python
data=data.join(one_hot["A"])
data=data.join(one_hot["B"])
data=data.join(one_hot["AB"])
data=data.join(one_hot["O"])
```
